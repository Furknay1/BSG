```python
import hashlib
import itertools
import string

# --- HEDEF ŞİFRELİ METİN ---
target_hash = "8bfa386a9a5a9a81af19837f2461b2e2"

print(f"Hedef Hash: {target_hash}")
print("Saldırı Başlatılıyor...\n")

# --- YÖNTEM 1: SÖZLÜK SALDIRISI (Dictionary Attack) ---
# Öğrencilerin sık kullandığı kelimelerden oluşan mini bir sözlük
common_passwords = [
    "123456", "password", "1234", "admin", "sifre", "deneme", "test",
    "merhaba", "hello", "root", "user", "qwerty", "12345", "istanbul",
    "turkiye", "bilgisayar", "proje", "odev", "kripto", "guvenlik"
]

print("--- YÖNTEM 1: Sözlük Saldırısı Deneniyor ---")
found = False

for word in common_passwords:
    # Kelimeyi MD5 ile şifrele
    hashed_word = hashlib.md5(word.encode('utf-8')).hexdigest()

    if hashed_word == target_hash:
        print(f"\n[!] ŞİFRE KIRILDI! (Sözlükte bulundu)")
        print(f"    Şifreli Metin: {target_hash}")
        print(f"    Çözülmüş Hali (Plaintext): {word}")
        found = True
        break

if not found:
    print("Sözlükte bulunamadı. Yöntem 2'ye geçiliyor...")

# --- YÖNTEM 2: BRUTE FORCE (Kaba Kuvvet) ---
# Eğer sözlükte yoksa, 1'den 5 karaktere kadar tüm kombinasyonları dener.
# UYARI: Karakter sayısı arttıkça süre çok uzar.
if not found:
    print("\n--- YÖNTEM 2: Brute Force Saldırısı (Max 5 Karakter) ---")
    charset = string.ascii_lowercase + string.digits # a-z ve 0-9 arasını dener

    # 1 karakterden 5 karaktere kadar uzunlukları dene
    for length in range(1, 6):
        print(f"{length} haneli şifreler deneniyor...")
        for attempt in itertools.product(charset, repeat=length):
            attempt_word = "".join(attempt)
            hashed_attempt = hashlib.md5(attempt_word.encode('utf-8')).hexdigest()

            if hashed_attempt == target_hash:
                print(f"\n[!] ŞİFRE KIRILDI! (Brute Force ile bulundu)")
                print(f"    Şifreli Metin: {target_hash}")
                print(f"    Çözülmüş Hali (Plaintext): {attempt_word}")
                found = True
                break
        if found:
            break

if not found:
    print("\n[-] Şifre bu basit tarama ile kırılamadı. Arkadaşın daha karmaşık veya tuzlanmış (salted) bir yöntem kullanmış olabilir.")
```
